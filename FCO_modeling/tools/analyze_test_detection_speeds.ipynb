```python
import pickle
import numpy as np
```

```python
with open('detection_comparison_results.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
# get the raytracing results
raytracing_results = results['raytracing']
mean_duration = np.mean(raytracing_results['duration']).item()
mean_duration_std = np.std(raytracing_results['duration']).item()
print(f'Raytracing mean duration: {mean_duration*1000:.2f} ± {mean_duration_std*1000:.2f} ms')

# additionally get duration per fco
per_fco_durations = [duration/len(fco) for duration, fco in zip(raytracing_results['duration'], raytracing_results['fcos'])]
print (per_fco_durations)
mean_per_fco_duration = np.mean(per_fco_durations).item()
mean_per_fco_duration_std = np.std(per_fco_durations).item()
print(f'Raytracing mean duration per FCO: {mean_per_fco_duration*1000:.2f} ± {mean_per_fco_duration_std*1000:.2f} ms')
```

```python
# get the nn results
nn_results = results['nn']
mean_duration = np.mean(nn_results['duration']).item()
mean_duration_std = np.std(nn_results['duration']).item()
print(f'NN mean duration: {mean_duration*1000:.2f} ± {mean_duration_std*1000:.2f} ms')

# additionally get duration per fco
per_fco_durations = [duration/len(fco) for duration, fco in zip(nn_results['duration'], nn_results['fcos'])]
print (per_fco_durations)
mean_per_fco_duration = np.mean(per_fco_durations).item()
mean_per_fco_duration_std = np.std(per_fco_durations).item()
print(f'NN mean duration per FCO: {mean_per_fco_duration*1000:.2f} ± {mean_per_fco_duration_std*1000:.2f} ms')
```
